```python
def find_threat_group(name, alias_dict):
    """
    This function takes a name (group name or alias) and returns the main group name
    along with all associated aliases.

    :param name: Name of the threat actor group or one of its aliases.
    :param alias_dict: Dictionary with group names as keys and lists of aliases as values.

    :return: A dictionary containing the main group name and its aliases, or an error message.
    """
    # Normalize the input name (case-insensitive search)
    name_lower = name.lower()

    # Iterate through the alias dictionary to find the matching group or alias
    for group, aliases in alias_dict.items():
        # Check if the name matches the group or any of its aliases
        if name_lower == group.lower() or name_lower in [alias.lower() for alias in aliases]:
            return {"group": group, "aliases": aliases}

    # If no match is found, return an error message
```
