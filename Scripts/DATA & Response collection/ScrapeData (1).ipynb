```python
# Remove duplicates by converting the list to a set
unique_urls0 = list(set(urls0))
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Initialize an empty list to hold DataFrames
df_list = []

def scrape_data(url):
    # Send a GET request to the webpage
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the webpage content
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract the title
        title = soup.find('h1').get_text(strip=True)

        # Extract the description
        description_div = soup.find('div', class_='description-body')
        description = description_div.get_text(separator=" ", strip=True) if description_div else "No description found."



        # Return the extracted data
        return {
            'Title': title,
            'Description': description

        }
    else:
        print(f"Failed to retrieve the webpage at {url}. Status code:", response.status_code)
        return None

# Scrape data from each URL and add to the list of DataFrames
for url in unique_urls0:
    data = scrape_data(url)
    if data:
        df_list.append(pd.DataFrame([data]))

# Concatenate all DataFrames in the list
df_test1 = pd.concat(df_list, ignore_index=True)

# Display the DataFrame
#print(df_test1)

# Optionally, save the DataFrame to a CSV file
df_test1.to_csv('mitre_attack_techniques.csv', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


def scrape_cwe_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract CWE ID and Name
    cwe_id_name = soup.find('h2').text.strip()

    # Extract relevant sections with checks
    description = soup.find('div', id='Description')
    description = description.text.strip() if description else ''



    return {
        'CWE ID and Name': cwe_id_name,
        'Description': description
    }

# Scrape data for all CWEs
cwe_data = []
for url in cwe_urls:
    data = scrape_cwe_data(url)
    cwe_data.append(data)

# Convert to DataFrame
cwe_df = pd.DataFrame(cwe_data)


```
