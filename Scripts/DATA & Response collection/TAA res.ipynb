```python
def promo (data):
    prompt = (
        f"{data['Prompt']}"
        f"{data['Text']}"

        )
    response = client.chat.completions.create(model='llama3-8b-8192', messages=[
      {
        'role': 'user',
        'content':prompt,
      },
    ])
    #print(response['message']['content'])
    return response.choices[0].message.content.strip()
```
